## **Sistema de Avaliação de Filmes**

O sistema implementado a seguir tem como objetivo definir um percentual de aceitação de um filme assistido.
 
 Este sistema foi desenvolvido por **Carlos Henrique Rici Conceição**, **Gabriel José Curvo Honda** e **Mateus Augusto Firmindo Ribeiro**


* **Entrada (antecedentes):**

> História do Filme
>>Universo ((intervalo de valores nítidos/crisp): 0 a 5)

>>Conjunto difuso (valores difusos): ruim, medio, bom

> Atuações
>>Universo ((intervalo de valores nítidos/crisp): 0 a 5)

>>Conjunto difuso (valores difusos): ruim, medio, bom

> Efeitos Especiais
>>Universo ((intervalo de valores nítidos/crisp): 0 a 5)

>>Conjunto difuso (valores difusos): ruim, medio, bom


* **Saída (consequentes):**

>Avaliação do Filme
>>Universo ((intervalo de valores nítidos/crisp): 0 a 5)

>>Conjunto difuso (valores difusos): muito ruim, ruim, mediano, bom, muito bom


##Adicionando as variáveis e Criando as Funções de Pertinência

```python
from skfuzzy import control as ctrl
import skfuzzy as fuzz
import numpy as np

# Declarando os antecedentes do problema
historia_do_filme = ctrl.Antecedent(np.arange(0, 6, 1), 'historia do filme')
atuacoes = ctrl.Antecedent(np.arange(0, 6, 1), 'atuacoes')
efeitos_especiais = ctrl.Antecedent(np.arange(0, 6, 1), 'efeitos especiais')

# Declarando as consequencias do problema
avaliacao_do_filme = ctrl.Consequent(np.arange(0, 6, 1), 'avaliacao do filme',  defuzzify_method='mom')

historia_do_filme.automf(names=['bom', 'medio','ruim'])
atuacoes.automf(names=['bom', 'medio','ruim'])
efeitos_especiais.automf(names=['bom', 'medio','ruim'])


historia_do_filme['ruim'] =  fuzz.trimf(historia_do_filme.universe, [0, 0, 2.5])
historia_do_filme['medio'] =  fuzz.trimf(historia_do_filme.universe, [1, 2.5, 4])
historia_do_filme['bom'] =  fuzz.trimf(historia_do_filme.universe, [2.5, 5, 5])

atuacoes['ruim'] =  fuzz.trimf(atuacoes.universe, [0, 0, 2.5])
atuacoes['medio'] =  fuzz.trimf(atuacoes.universe, [1, 2.5, 4])
atuacoes['bom'] =  fuzz.trimf(atuacoes.universe, [2.5, 5, 5])

efeitos_especiais['ruim'] =  fuzz.trimf(efeitos_especiais.universe, [0, 0, 2.5])
efeitos_especiais['medio'] =  fuzz.trimf(efeitos_especiais.universe, [1, 2.5, 4])
efeitos_especiais['bom'] =  fuzz.trimf(efeitos_especiais.universe, [2.5, 5, 5])

# Avaliação do filme
avaliacao_do_filme.automf(names=['muito ruim', 'ruim', 'medio', 'bom', 'muito bom'])
```

```python
historia_do_filme.view()
efeitos_especiais.view()
atuacoes.view()
avaliacao_do_filme.view()
```

##Criando as Regras de Decisões Difusas

```python
rule1 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['ruim'] & efeitos_especiais['ruim'], avaliacao_do_filme['muito ruim'])
rule2 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['ruim'] & efeitos_especiais['medio'], avaliacao_do_filme['ruim'])
rule3 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['ruim'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule4 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['medio'] & efeitos_especiais['ruim'], avaliacao_do_filme['ruim'])
rule5 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['medio'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule6 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['medio'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule7 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['bom'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule8 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['bom'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule9 = ctrl.Rule(historia_do_filme['ruim'] & atuacoes['bom'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule10 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['ruim'] & efeitos_especiais['ruim'], avaliacao_do_filme['ruim'])
rule11 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['ruim'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule12 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['ruim'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule13 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['medio'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule14 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['medio'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule15 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['medio'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule16 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['bom'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule17 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['bom'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule18 = ctrl.Rule(historia_do_filme['medio'] & atuacoes['bom'] & efeitos_especiais['bom'], avaliacao_do_filme['bom'])
rule19 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['ruim'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule20 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['ruim'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule21 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['ruim'] & efeitos_especiais['bom'], avaliacao_do_filme['medio'])
rule22 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['medio'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule23 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['medio'] & efeitos_especiais['medio'], avaliacao_do_filme['medio'])
rule24 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['medio'] & efeitos_especiais['bom'], avaliacao_do_filme['bom'])
rule25 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['bom'] & efeitos_especiais['ruim'], avaliacao_do_filme['medio'])
rule26 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['bom'] & efeitos_especiais['medio'], avaliacao_do_filme['bom'])
rule27 = ctrl.Rule(historia_do_filme['bom'] & atuacoes['bom'] & efeitos_especiais['bom'], avaliacao_do_filme['muito bom'])



```

##Criando e Simulando um Controle Avaliativo

```python


avaliacao_do_filme_ctrl = ctrl.ControlSystem([
                                              rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8,
                                              rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16,
                                              rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24,
                                              rule25, rule26, rule27
                                              ])

avaliacao_do_filme_simulador = ctrl.ControlSystemSimulation(avaliacao_do_filme_ctrl)

# Entrando com alguns valores avaliação do filme
avaliacao_do_filme_simulador.input['historia do filme'] = 1
avaliacao_do_filme_simulador.input['atuacoes'] = 1
avaliacao_do_filme_simulador.input['efeitos especiais'] = 5

# Computando o resultados
avaliacao_do_filme_simulador.compute()
print("Avaliação do filme:", avaliacao_do_filme_simulador.output['avaliacao do filme'])

```

```python
historia_do_filme.view(sim=avaliacao_do_filme_simulador)
efeitos_especiais.view(sim=avaliacao_do_filme_simulador)
atuacoes.view(sim=avaliacao_do_filme_simulador)
```
